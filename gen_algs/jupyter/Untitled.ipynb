```python
from tensorflow import keras
```

```python
input_neurons = 12
inner_neurons = 3
output_neurons = 4
```

```python
input_layer = keras.Input(shape=(input_neurons,), dtype='int32', name='sensory')
inner_layer = keras.layers.Dense(inner_neurons, activation='tanh', name='internal')(input_layer)
output = keras.layers.Dense(output_neurons, activation='tanh', name='output')(inner_layer)
model = keras.Model(inputs=[input_layer], outputs=[output])
```

```python
keras.utils.plot_model(model, show_shapes=True)
```
